```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir una señal de entrada de 10 a 20 valores
signal = [2, 5, 1, 8, 7, 6, 2, 4, 3, 5, 7, 6, 8, 5, 3]

# 2. Definir un kernel de tamaño 3
kernel = [1, 0, -1]  # También puedes probar con [1/3, 1/3, 1/3]

# 3. Implementar la función de convolución 1D
def convolution_1d(signal, kernel):
    n = len(signal)
    k = len(kernel)
    pad = k // 2
    # Pad the signal with zeros at el inicio y final
    padded_signal = [0]*pad + signal + [0]*pad
    result = []
    for i in range(n):
        acc = 0
        for j in range(k):
            acc += padded_signal[i + j] * kernel[j]
        result.append(acc)
    return result

# 4. Calcular la convolución
conv_result = convolution_1d(signal, kernel)

# 5. Graficar la señal, el kernel y el resultado
plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.stem(signal)
plt.title('Señal de Entrada')
plt.xlabel('Índice')
plt.ylabel('Valor')

plt.subplot(1,3,2)
plt.stem(kernel)
plt.title('Kernel')
plt.xlabel('Índice')
plt.ylabel('Valor')

plt.subplot(1,3,3)
plt.stem(conv_result)
plt.title('Resultado de la Convolución')
plt.xlabel('Índice')
plt.ylabel('Valor')

plt.tight_layout()
plt.show()
```
